```python
#science time is a nominal variable , we will use label encoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['time']=encoder.fit_transform(df['time'])
```

```python
x=df.drop('time',axis=1)
```

```python
y=df['time']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
#handling the missing value
#data encoding
#feature scaling
```

```python
from sklearn.impute import SimpleImputer # for missing value
```

```python
from sklearn.preprocessing import OneHotEncoder # for encoding
```

```python
from sklearn.preprocessing import StandardScaler # for scaling
```

```python
from sklearn.pipeline import Pipeline # a seq of data transformer # step by step missing value , encoding .... kadega
from sklearn.compose import ColumnTransformer# grp each pipeline steps on each columns
```

```python
cat_cols=['sex','smoker','day']
num_cols=['total_bill','tip','size']
```

```python
#feature engineering automationusing pipeline and columntransformer
num_pipeline=Pipeline(steps=[('imputation',SimpleImputer(strategy='median')),('scaling',StandardScaler())])
```

```python
cat_pipeline=Pipeline(steps=[('imputation',SimpleImputer(strategy='most_frequent')),('encoding',OneHotEncoder())])
```

```python
preProcessor=ColumnTransformer([("num_pipeline",num_pipeline,num_cols),('cat_pipelines',cat_pipeline,cat_cols)])
```

```python
x_train=preProcessor.fit_transform(x_train)
x_test=preProcessor.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
## eg
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
models={"support vector classifier": SVC(),"DT classifier": DecisionTreeClassifier(),"Logistic regression":LogisticRegression()}
```

```python
from sklearn.metrics import accuracy_score

def model_train_eval(x_train, y_train, x_test, y_test, models):
    evaluation = {}
    for i in range(len(models)):
        model = list(models.values())[i]
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        model_score = accuracy_score(y_test, y_pred)
        evaluation[list(models.keys())[i]] = model_score
    return evaluation
```

```python
model_train_eval(x_train,y_train,x_test,y_test,models)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf
```

```python
#hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
params={
    'max_depth':[1,2,3,5,10,None],
    'n_estimators':[50,100,200,300],
    'criterion':['gini','entropy']
}
```

```python
params
```

```python
clf=RandomizedSearchCV(rf,param_distributions=params,cv=5,verbose=3,scoring='accuracy')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```
